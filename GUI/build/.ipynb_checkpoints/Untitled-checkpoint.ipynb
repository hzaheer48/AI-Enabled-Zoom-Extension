```python
import math
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler 
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import re
import joblib
```

```python
data = pd.read_csv("bps.csv")
indices_to_select = list(range(0, len(data), 10))  # Starting from 0, incrementing by 10
selected_rows = data.iloc[indices_to_select]
data = selected_rows
```

```python
data.shape
```

```python
data.head(20)
```

```python
def extract_units(text):
    units = re.findall(r'\b[a-zA-Z]+(?:/[a-zA-Z]+)?\b', text)
    return units
data['units_ip'] = data['ip_len'].apply(extract_units)
```

```python
unique_units = data['units_ip'].explode().unique()
print(unique_units)
```

```python
conversion_factors_bits = {
    'b/s': 0.000001,
    'kb/s': 0.001,
    'Mb/s': 1,
    'Gb/s': 1000
}
```

```python
def conversion(count_str):
    extractor = count_str.split()
    if('b/s' in extractor[1]):
        value = int(extractor[0]) * conversion_factors_bits[extractor[1]]
    return value
data['converted_bits'] = data['ip_len'].apply(conversion)
```

```python
first_500_rows = data['converted_bits']
data['converted_bits'] = first_500_rows
```

```python
columns_to_keep = ['converted_bits']
data = data[columns_to_keep]
print(data)
```

```python
converted_bits = data['converted_bits']
values = converted_bits.values
training_data_len = math.ceil(len(values)* 0.8)

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(values.reshape(-1,1))
train_data = scaled_data[0: training_data_len, :]

x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
    
x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))

scaler_filename = "scaler.pkl"
joblib.dump(scaler, scaler_filename)
```

```python
test_data = scaled_data[training_data_len-60: , : ]
x_test = []
y_test = values[training_data_len:]

for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])
x_test = np.array(x_test)
print(x_test[0])
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
print(x_test[0])
```

```python
print(x_test)
```

```python
model = keras.Sequential()
model.add(layers.LSTM(128, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(layers.LSTM(64, return_sequences=False))
model.add(layers.Dense(32))
model.add(layers.Dense(16))
model.add(layers.Dense(8))
model.add(layers.Dense(4))
model.add(layers.Dense(2))
model.add(layers.Dense(1))
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, batch_size= 16, epochs=100)
```

```python
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)
rmse = np.sqrt(np.mean(predictions - y_test)**2)
rmse
```

```python
data = data.filter(['converted_bits'])
train = data[:training_data_len]
validation = data[training_data_len:]
validation['Predictions'] = predictions
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Date')
plt.ylabel('Bandwidth')
plt.plot(train)
plt.plot(validation[['converted_bits', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```

```python
model.save('trained_model.h5')
print("Model saved successfully.")
```
